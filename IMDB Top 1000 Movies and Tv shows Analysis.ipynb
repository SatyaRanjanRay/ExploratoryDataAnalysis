## Step4: Exploratory Analysis

### Top 10 Genres

```python
from collections import Counter
genre=[]
for x in df['Genre']:
    for y in x.split(','):
        genre.append(y.strip().lower())

count=Counter(genre)
count=count.most_common()[:10]
x,y=map(list,zip(*count))

fig,axs=plt.subplots(figsize=(20,5))
p=sns.barplot(x=y,y=x)
p.set_ylabel("Genres", weight = "bold")
p.set_title("Top 10 Genres",fontdict={"size":15}, weight = "bold")
plt.show()
```

### Runtime of the Movies

```python
import re

duration = []

# Looping through each runtime value in the data
for x in df["Runtime"]:
    # Compiling a regular expression pattern to match digits
    p = re.compile('\d+')
    # Finding all occurrences of digits in the runtime value
    j = p.findall(x)
    
    # calculate the duration if there are two sets of digits
    if len(j) == 2:
        duration.append(int(j[1]) - int(j[0]) + 1)
    # If there is only one set of digits then use that as the duration
    else:
        duration.append(int(j[0]) + 1)

# Adding the duration list as a new column in the data
df['duration'] = duration
df['duration'].head()
```

```python
fig,axs=plt.subplots(figsize=(20,5))
g=sns.kdeplot(df['duration'])
g.set_title("Time Duration of movies", weight = "bold")
```
